–î–æ–º–∞—à–Ω—ñ –∑–∞–≤–¥–∞–Ω–Ω—è –Ω–∞ –≤—ñ–¥–ø—Ä–∞—Ü—é–≤–∞–Ω–Ω—è –ø–æ–Ω—è—Ç—Ç—è —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó
–ó–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü—ñ—è –∑–∞–ª–µ–∂–Ω–æ—Å—Ç—ñ –æ—Ü—ñ–Ω–∫–∏ –≤—ñ–¥ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω, –≤–∏—Ç—Ä–∞—á–µ–Ω–∏—Ö –Ω–∞ –Ω–∞–≤—á–∞–Ω–Ω—è:  ùëì(ùë•)=10ùë•‚àí5 .
–ó –¥–æ–ø–æ–º–æ–≥–æ—é Python –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó.
–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É–π—Ç–µ –±—ñ–±–ª—ñ–æ—Ç–µ–∫—É matplotlib –¥–ª—è –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –≤ –¥—ñ–∞–ø–∞–∑–æ–Ω—ñ –∑–Ω–∞—á–µ–Ω—å –≤—ñ–¥ 0 –¥–æ 10 –≥–æ–¥–∏–Ω. –ü—Ä–æ–∞–Ω–∞–ª—ñ–∑—É–π—Ç–µ, —è–∫ –∑–º—ñ–Ω–∞ –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≥–æ–¥–∏–Ω –≤–∏–≤—á–µ–Ω–Ω—è –≤–ø–ª–∏–≤–∞—î –Ω–∞ –æ—Ü—ñ–Ω–∫—É.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f_lin = 10 * x - 5
derivative_f_lin = sp.diff(f_lin, x)
derivative_f_lin
```

```python
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='–û—Ü—ñ–Ω–∫–∞ f(x) = 10 * x - 5', color='blue')
plt.plot(x, df, label="–í–∏—Ç—Ä–∞—á–µ–Ω–∏–π —á–∞—Å f'(x) = 10", color='green')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

–í–∏—Å–Ω–æ–≤–æ–∫: —â–∏–º –±—ñ–ª—å—à–µ –∑–∞—Ç—Ä–∞—á–µ–Ω–æ —á–∞—Å—É, —Ç–∏–º –≤–∏—â–µ –æ—Ü—ñ–Ω–∫–∞.

–í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—é—á–∏ Python —ñ SymPy, –æ–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—ñ –¥–ª—è —Ñ—É–Ω–∫—Ü—ñ–π  ùëì(ùë•)=sin(ùë•)  —Ç–∞  ùëì(ùë•)=ln(ùë•) . –ó–æ–±—Ä–∞–∑—ñ—Ç—å —Ü—ñ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—Ö–Ω—ñ –ø–æ—Ö—ñ–¥–Ω—ñ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é matplotlib.

```python
x = sp.symbols('x')
f1 = sp.sin(x)
f2 = sp.ln(x)
derivative_f1 = sp.diff(f1, x)
derivative_f2 = sp.diff(f2, x)
derivative_f1, derivative_f2
```

```python
x_sin = np.linspace(-10, 10, 100)
x_ln = np.linspace(0.1, 10, 100)


# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x_sin, np.sin(x_sin), label='f(x ) = sin(x)', color='blue')
plt.plot(x_sin, np.cos(x_sin), label="f'(x) = cos(x)", color='green')

plt.plot(x_ln, np.log(x_ln), label='f(x ) = ln(x)', color='yellow')
plt.plot(x_ln, 1/x_ln, label="f'(x) = 1/x", color='red')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

–ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ –∫–æ–º–ø–∞–Ω—ñ—ó –≤—ñ–¥ –æ–±—Å—è–≥—É –≤–∏—Ä–æ–±–Ω–∏—Ü—Ç–≤–∞ –∑–∞–¥–∞–Ω—ñ —Ñ—É–Ω–∫—Ü—ñ—î—é  ùëì(ùë•)=ùë•2‚àí5ùë•+1 , –¥–µ  ùë•  ‚Äî –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏—Ä–æ–±–ª–µ–Ω–∏—Ö –æ–¥–∏–Ω–∏—Ü—å. –û–±—á–∏—Å–ª—ñ—Ç—å –ø–æ—Ö—ñ–¥–Ω—É —Ü—ñ—î—ó —Ñ—É–Ω–∫—Ü—ñ—ó –∑ –¥–æ–ø–æ–º–æ–≥–æ—é SymPy —Ç–∞ –≤—ñ–∑—É–∞–ª—ñ–∑—É–π—Ç–µ –∑–Ω–∞—á–µ–Ω–Ω—è —Ñ—É–Ω–∫—Ü—ñ—ó —ñ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó –Ω–∞ –≤—ñ–¥—Ä—ñ–∑–∫—É –≤—ñ–¥ 0 –¥–æ 100.

```python
x = sp.symbols('x')
f_x = x ** 2 - 5 * x + 1
derivative_f_x = sp.diff(f_x, x)
derivative_f_x
```

```python
x = np.linspace(0, 10, 100)
f_x = x ** 2 - 5 * x + 1
df = 2 * x - 5

# –°—Ç–≤–æ—Ä–µ–Ω–Ω—è –≥—Ä–∞—Ñ—ñ–∫—É
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='–ó–∞–≥–∞–ª—å–Ω—ñ –≤–∏—Ç—Ä–∞—Ç–∏ f(x) = x2 - 5x + 1', color='blue')
plt.plot(x, df, label="–ó–º—ñ–Ω–∞ –≤–∏—Ç—Ä–∞—Ç –ø—Ä–∏ +1 –æ–¥–∏–Ω–∏—Ü—ñ f'(x) = 2x - 5", color='green')

# –î–æ–¥–∞–≤–∞–Ω–Ω—è –ª–µ–≥–µ–Ω–¥–∏, –º—ñ—Ç–æ–∫ —Ç–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞
plt.title("–ì—Ä–∞—Ñ—ñ–∫ —Ñ—É–Ω–∫—Ü—ñ—ó —Ç–∞ —ó—ó –ø–æ—Ö—ñ–¥–Ω–æ—ó")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
