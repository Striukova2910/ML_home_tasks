ДЗ 4. Одновимірний аналіз категоріальної колонки
За наведеним прикладом вище, проведіть аналіз залежності між категоріальною колонкою і цільовою для колонок

Чи є вплив цих змінних на цільову та який саме?

```python
def bi_cat_countplot(df, column, hue_column):
    unique_hue_values = df[hue_column].unique()
    fig, axes = plt.subplots(nrows=1, ncols=2)
    fig.set_size_inches(14,6)

    pltname = f'Нормалізований розподіл значень за категорією: {column}'
    proportions = df.groupby(hue_column)[column].value_counts(normalize=True)
    proportions = (proportions*100).round(2)
    ax = proportions.unstack(hue_column).sort_values(
        by=unique_hue_values[0], ascending=False
        ).plot.bar(ax=axes[0], title=pltname)


    # % розподіл
    proportions = df.groupby(hue_column)[column].value_counts(normalize=True)
    proportions = (proportions * 100).round(2)
    proportions.unstack(hue_column).plot.bar(
        ax=axes[0],
        title=f'Нормалізований розподіл значень за категорією: {column}'
    )

    # кількість
    counts = df.groupby(hue_column)[column].value_counts()
    counts.unstack(hue_column).plot.bar(
        ax=axes[1],
        title=f'Кількість даних за категорією: {column}'
    )

    plt.show()
```

```python
def uni_cat_target_compare(df, column):
    bi_cat_countplot(df, column, hue_column='TARGET')
```

```python
df['CODE_GENDER'].value_counts()
```

```python
(df['CODE_GENDER'].value_counts(normalize=True) * 100).round(2)
```

```python
uni_cat_target_compare(df, 'CODE_GENDER')
```

 На нормалізованому графіку видно, що є різниця між Таргет = 1 і Таргет =0. Серед тих, хто протермінував (Таргет = 1), чоловіків більше, жінки переважають у Таргет = 0 - без протермінувань.
Висновок: колонка "CODE_GENDER" не сильно пов'язана з цільвою змінною, не є вирішальною, хоч і демонструє цікаву інформацію по статі, хто більше протерміновує оплату.

Зробіть висновок про наявність кореляції між змінними.

```python
df_pair = df[['AMT_ANNUITY', 'AMT_CREDIT']]
```

```python
df_pair.describe()
```

```python
def outlier_range(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    upper_bound = Q3 + 1.5 * IQR
    return upper_bound
```

```python
max_annuity = outlier_range(df_pair['AMT_ANNUITY'])
max_credit = outlier_range(df_pair['AMT_CREDIT'])
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x = df_pair['AMT_ANNUITY'], y = df_pair['AMT_CREDIT'])
plt.title('AMT_CREDIT vs AMT_ANNUITY with outliners')
plt.ticklabel_format(axis='both', style='plain')
```

```python
df_pair_no_outliners = df_pair[(df_pair['AMT_ANNUITY'] < max_annuity) & (df_pair['AMT_CREDIT'] < max_credit)]
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x = df_pair_no_outliners['AMT_CREDIT'], y = df_pair_no_outliners['AMT_ANNUITY'])
plt.title('AMT_CREDIT vs AMT_ANNUITY without outliners')
plt.ticklabel_format(axis='both', style='plain')
plt.show()
```

Висновок: отримані графіки показують, що AMT_ANNUITY залежить від AMT_CREDIT, а видалення викидів дозволило краще побачити цю залежність.
