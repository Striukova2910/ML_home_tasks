```python
# dinner_bill > 30
dinner_bill_30 = df[(df['time'] == 'Dinner') & (df['total_bill'] > 30)]
print(f'The total amount of dinner bills > $30 is: ', len(dinner_bill_30)) # total amount dinner bills > 30
display(dinner_bill_30.head()) # to show head() 5 dinner bills > 30
```

4. Описова статистика
Обчисліть (кожен пункт оцінюється в 0.5 бала):
Середній розмір рахунку на всьому наборі даних
Максимальні чайові
Кількість пропущених значень у всіх колонках
Скільки в наборі даних рахунків за обіди і за вечері

```python
# mean bill
mean_bill = df['total_bill'].mean()
print(f'Mean bill is $:', round(mean_bill)) # before round() the mean amount was 19.78594...
```

```python
# max tips
max_tips = df['tip'].max()
print(f'Max tips is $:', max_tips)
```

```python
# Number of missing values in all columns
missing_values = df.isnull().sum()
print(missing_values)
print(f'THERE ARE NO MISSING VALUES')
```

```python
# Lunch and dinner bills amount in dataset
lunch_and_dinner = df['time'].value_counts()
print(lunch_and_dinner)
```

```python
# Mean bill by sex
mean_bill_by_sex = df.groupby('sex')['total_bill'].mean()
print(f'The mean bill by sex is:\n ', round(mean_bill_by_sex))
```

```python
# Mean tip by day
mean_tip_by_day = df.groupby('day')['tip'].mean()
print(f'The mean tip by day is:\n ', mean_tip_by_day)
# In this case round() makes the same result for all days.
# Probably it's better to use a previous form to see the difference
print("------------------------") # visual divider
print(f'The mean tip by day is:\n ', round(mean_tip_by_day))
```

```python
# Daily visitors amount
daily_visits = df.groupby('day')['size'].sum()
print(f'The daily number of visits is:\n ', daily_visits)
```

6. Створення нової колонки

7. Сортування
Відсортуйте датасет і виведіть результат сортування в кожному випадку окремо (кожен пункт оцінюється в 0.5 бала):

За розміром рахунку (зростання)
За чайовими (спадання)
Додатково виведіть топ-5 найбільших чайових використовуючи метод nlargest.

```python
# sorted by bill - increasing
sorted_by_bill =df.sort_values(['total_bill'], ascending=[True])
display(sorted_by_bill[['total_bill']].head(10))
```

```python
# sorted by tips --- decreasing + top 5 tips
sorted_by_tips =df.sort_values(['tip'], ascending=[False])
top_tips = df.nlargest(5, ['tip'])
display(sorted_by_tips[['tip']].head(10), f'Top 5 tips:', top_tips[['tip']])
```
